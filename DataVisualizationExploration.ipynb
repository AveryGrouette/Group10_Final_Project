1. Import libraries

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('clean_data.csv')
df.head(5)
```

```python
if 'starttime' < '2013-07-01 00:00:00':
    day_station_number = df.groupby([df['starttime'].str[:-8], 'ss_id'])['ss_id'].count()
# Set the figure size

day_station_number.unstack().plot(kind = 'barh',figsize=(10,200), width =0.8)

plt.title('Number of Trips for Each Station for a Day')
plt.xlabel('Day')
plt.ylabel('Number of Trips')
plt.legend(ncol = 5, loc='center left', bbox_to_anchor=(1.5,0.5))

plt.show()
```

```python
# Average Duration of The Trip for Each Station

average_data = df.groupby(['ss_id'])['tripduration'].mean()

# Set the figure size
plt.figure(figsize=(15, 50))  # Width: 12 inches, Height: 6 inches


# Create the bar graph
average_data.plot(kind='barh')

# Set other properties for the plot
plt.title('Average Duration of The Trip for Each Station')
plt.xlabel('Group of Station')
# plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees

plt.ylabel('Average duration')
plt.legend(ncol=10, loc='lower center', bbox_to_anchor=(0.5, 1.2))

# Show the plot
plt.show()
```
