1. Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_data.csv')
df.head(5)
```

```python
day_station_number = df.groupby([df['starttime'].str[:-9]])['ss_id'].count()
# Set the figure size

day_station_number.plot(kind = 'line',figsize=(30,10))

plt.title('Number of Trips for  for a Day')
plt.xlabel('Day')
plt.ylabel('Number of Trips')

plt.show()
```

```python
# Average Duration of The Trip for Each Station

average_data = df.groupby(['ss_id'])['tripduration'].mean()

# Set the figure size
# plt.figure(figsize=(15, 50))  # Width: 12 inches, Height: 6 inches


# Create the bar graph
average_data.plot(kind='barh',figsize=(20,70))

# Set other properties for the plot
plt.title('Average Duration of The Trip for Each Station')
plt.xlabel('Average duration')
# plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees

plt.ylabel('Station ID')
# plt.legend(ncol=10, loc='lower center', bbox_to_anchor=(1.3, 0.2))

# Show the plot
plt.show()
```

```python
max_min = df.groupby('ss_id')['tripduration'].agg(['max', 'min'])
max_min.sort_values(by=['max','min'], ascending=False)
```
